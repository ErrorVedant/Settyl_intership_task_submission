```python
import re
import pandas as pd
def clean(text):
  lowercase_text = text.lower()
  cleaned_string = re.sub(r'\[.*?\]|\(.*?\)|\'.*?\'|[0-9]{3}[a-z]', '', lowercase_text)
  return cleaned_string

result = map(clean, df.externalStatus)
result_list = list(result)

new_df = pd.DataFrame({'externalStatus': result_list , 'internalStatus': df.internalStatus})
```

```python
from sklearn.model_selection import train_test_split
X=new_df.externalStatus
y=new_df.internalStatus
X_train, X_test, y_train, y_test = train_test_split(result_list, df.internalStatus, test_size=0.3, random_state=42)
```

```python

import tensorflow as tf
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
out_X_train = vectorizer.fit_transform(X_train)
features_encoded=out_X_train.toarray()  # Convert the sparse matrix to a dense matrix for printing
features=vectorizer.get_feature_names_out()  # Get the names of the features (tokens)
```

```python
from pydantic import BaseModel
class ResponseQuery(BaseModel):
    externalStatus: str

from fastapi import FastAPI, Request, HTTPException
from fastapi.responses import JSONResponse
import pickle
import numpy as np

with open('model.pkl', 'rb') as f:
    model = pickle.load(f)

with open('features.pkl', 'rb') as f:
    features = pickle.load(f)

app = FastAPI()

@app.post("/predict")
async def predict(data:ResponseQuery):
    data = data.dict()
    text = data["externalStatus"]
    if text is None:
        raise HTTPException(status_code=400, detail="Missing text input")

    def clean(text):
        lowercase_text = text.lower()
        cleaned_string = re.sub(r'\[.*?\]|\(.*?\)|\'.*?\'|[0-9]{3}[a-z]', '', lowercase_text)
        return cleaned_string

    cleaned_text = clean(text)

    def match_label(given_str, features_depends):
        a = [0] * len(features_depends)  # Initialize with zeros
        for i in range(len(features_depends)):
            if features_depends[i] in given_str:
                a[i]=1
        return a

    X_test_array = match_label(cleaned_text, features)
    X_test_array = np.array([X_test_array])

    prediction = model.predict(X_test_array)

    num_rows, num_cols = prediction.shape
    y_pred = [None] * num_rows
    for i in range(num_rows):
        max_index = 0
        for k in range(1, num_cols):
            if prediction[i][k] > prediction[i][max_index]:
                max_index = k
        for k in range(num_cols):
            if k == max_index:
                y_pred[i]=column_names_y_train[max_index]

    return JSONResponse({"prediction": y_pred[0]})
```
